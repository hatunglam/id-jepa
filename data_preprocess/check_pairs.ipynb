```python
import matplotlib.pyplot as plt
import cv2
import random
from pathlib import Path
```

```python
DATA_ROOT = Path("D:/Documents_D/processed/train")  
N_SAMPLES = 10  
```

```python
rgb_dir = DATA_ROOT / "rgb"
depth_dir = DATA_ROOT / "depth"

rgb_files = sorted(rgb_dir.glob("*.png"))
depth_files = sorted(depth_dir.glob("*.png"))

assert len(rgb_files) == len(depth_files)
total = len(rgb_files)


indices = random.sample(range(total), N_SAMPLES)

for idx in indices:
    rgb_path = rgb_files[idx]
    depth_path = depth_files[idx]

    rgb = cv2.imread(str(rgb_path))
    rgb = cv2.cvtColor(rgb, cv2.COLOR_BGR2RGB)

    depth = cv2.imread(str(depth_path), cv2.IMREAD_UNCHANGED)

    plt.figure(figsize=(10, 4))
    
    plt.subplot(1, 2, 1)
    plt.title(f"RGB [{rgb_path.name}]")
    plt.imshow(rgb)
    plt.axis("off")

    plt.subplot(1, 2, 2)
    plt.title(f"Depth [{depth_path.name}]")
    plt.imshow(depth, cmap="gray")
    plt.axis("off")

    plt.show()
```
